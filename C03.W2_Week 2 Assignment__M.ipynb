```python
# Install the packages we need

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data

NBA19 = pd.read_excel("Assignment Data/NBA2019odds.xlsx")
NBA19
```

```python
# 2.	Calculate the implied win probability based on the bookmaker odds
```

```python
# Calculate the implied win probability from the decimal odds

NBA19['winprob']= 1/(NBA19['winodds'])/(1/(NBA19['winodds'])+ 1/(NBA19['loseodds']))
```

```python
# 3.	Drop from the data all games for which home = 0
```

```python
NBA19 = NBA19[NBA19.home == 1]
NBA19
```

```python
np.corrcoef(NBA19["winprob"], NBA19['win'])
```

```python
# Q1: What is the correlation between the home team win probability and home team wins across the entire 2018/19 season? 
# A1: +0.397
```

```python
# 4.	Then create the following subsets 
#     a. One for games where the absolute value of the points difference is less than 9 and another for games where the points difference is greater than 9 (9 is the median points difference in the data) 
#     b. One for games that went to overtime, and one for games which finished in regulation time 
#     c. One for games played in calendar year 2018, one for games played in calendar year 2019 
#     d. One for each month in the data (October to April)
```

```python
# a. One for games where the absolute value of the points difference is less than 9 and another for games where the points difference is greater than 9 (9 is the median points difference in the data) 
NBA19['absdiff'] = abs(NBA19['teampts'] - NBA19['oppopts']) 
NBA19
```

```python
np.median(NBA19['absdiff'])
```

```python
NBAl = NBA19[NBA19['absdiff']<9]
np.corrcoef(NBAl["winprob"], NBAl['win'])
```
