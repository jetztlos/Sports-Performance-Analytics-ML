# I. Coding Run Expectancy Dataset (2017)

```python
#Your Code Here
```

```python
# 1.	Import pandas and numpy packages and read in “MLBAM17.csv” data
```

```python
# Read in MLBAM Data for 2017

MLBAM17 = pd.read_csv("../Data/MLBAM17.csv")
```

```python
display(MLBAM17)
```

```python
# 2.	Keep only the relevant columns (see exercise notebook)
```

```python
# Drop column with no relevant information
MLBAM17.drop(['Unnamed: 0'], axis=1, inplace=True)
MLBAM17
```

```python
# Keep relevant columns
RE17 = MLBAM17[['batterName','batterId','event', 'start1B', 'start2B', 'start3B', 'end1B', 'end2B', 'end3B',\
                'startOuts','endOuts','runsFuture','runsOnPlay','outsInInning',\
                'stand', 'throws','venueId', 'stadium', 'batterPos']].copy()

display(RE17)
```

```python
# 3.	Create three indicator variables (one for each base) to denote whether or not each base is occupied prior to the plate appearance
```

```python
# Create indicator variables to determine whether base is occupied prior to plate appearance
RE17['Start1'] = np.where(pd.isnull(RE17['start1B']),0,1)
RE17['Start2'] = np.where(pd.isnull(RE17['start2B']),0,1)
RE17['Start3'] = np.where(pd.isnull(RE17['start3B']),0,1)
```

```python
# 4.	Create a “start state” variable to denote the base out state prior to each plate appearance
```

```python
# Create starting base out state for plate appearance
RE17['Start_State'] = (RE17['Start1'].astype(str) + RE17['Start2'].astype(str) + RE17['Start3'].astype(str)+\
                       " " + RE17['startOuts'].astype(str))
RE17
```

```python
# 5.	Create three indicator variables (one for each base) to denote whether or not each base is occupied after the plate appearance
```

```python
# Create indicator variables to determine whether base is occupied after plate appearance
RE17['End1'] = np.where(pd.isnull(RE17['end1B']),0,1)
RE17['End2'] = np.where(pd.isnull(RE17['end2B']),0,1)
RE17['End3'] = np.where(pd.isnull(RE17['end3B']),0,1)
```

```python
# 6.	Create an “end state” variable to denote the base out state after each plate appearance
```

```python
# Create ending base out state for plate appearance
RE17['End_State'] = (RE17['End1'].astype(str) + RE17['End2'].astype(str) + RE17['End3'].astype(str)+\
                     " " + RE17['endOuts'].astype(str))
RE17
```

```python
# 7.	Restrict data to plays where either there is a change in state/runs scored and innings in which there were exactly 3 outs recorded
```

```python
# Restrict data to plays where either there is a change in state or runs scored and innings that had exactly 3 outs
RE17 = RE17[((RE17.Start_State != RE17.End_State) | (RE17.runsOnPlay > 0)) & (RE17.outsInInning == 3)]
display(RE17)
```

```python
# 8.	Calculate run expectancy by starting state and merge this into play by play data
```

```python
# Calculate run expectancy by starting state and merge into play by play data
Start_RunExp = RE17.groupby(['Start_State'])['runsFuture'].mean().reset_index().rename(columns={'runsFuture':'Start_RE'})
RE17 = pd.merge(RE17, Start_RunExp, on=['Start_State'], how='left')
display(RE17)
```

```python
# 9.	Using the run expectancy data by starting state, create a variable to denote run expectancy by ending state 
# (in addition, make sure to include the base out states for which there are 3 outs with run expectancies of 0 to the run expectancy by ending state data)
```

```python
# Code run expectancy values by end state using start state run expectancy including end states with 3 outs
Base_State_3 = [pd.Series(['000 3', 0], index=Start_RunExp.columns),
                pd.Series(['001 3', 0], index=Start_RunExp.columns),
                pd.Series(['010 3', 0], index=Start_RunExp.columns),
                pd.Series(['011 3', 0], index=Start_RunExp.columns),
                pd.Series(['100 3', 0], index=Start_RunExp.columns),
                pd.Series(['101 3', 0], index=Start_RunExp.columns),
                pd.Series(['110 3', 0], index=Start_RunExp.columns),
                pd.Series(['111 3', 0], index=Start_RunExp.columns)]

Start_RunExp = Start_RunExp.append(Base_State_3, ignore_index=True)

End_RunExp = Start_RunExp.rename(columns={'Start_State':'End_State', 'Start_RE':'End_RE'})

display(End_RunExp)
```

```python
# 10.	Merge the ending state run expectancy into the play by play data
```

```python
# Merge into play by play data
RE17 = pd.merge(RE17, End_RunExp, on=['End_State'], how='left')
display(RE17)
```

```python
# 11.	Calculate the run value of each event
```

```python
# Run Value of each event
RE17['Run_Value'] = RE17['runsOnPlay'] + RE17['End_RE'] - RE17['Start_RE']

display(RE17)
```

```python
# Q1: What percent of plate appearances resulted in fly outs in 2017?
RE17['Count'] = 1
Flyout = RE17[RE17['event']=='Flyout']
sum(Flyout['Count'])/sum(RE17['Count'])
```

```python
# Q2: How many plate appearances had a starting base state in which the bases were loaded (all bases were occupied)?
RE17_BLoad = RE17[(RE17['Start_State'] == "111 0")|(RE17['Start_State'] == "111 1")|(RE17['Start_State'] == "111 2")]
RE17_BLoad
```

```python
# Q3: Calculate aggregate strikeouts by player position (that is, aggregate (sum) data at the positional level and not the player level).  
# What was the highest aggregate strikeout total by position?
Strikeouts = RE17[RE17['event'] == "Strikeout"]
Strikeouts.groupby(['batterPos'])['Count'].sum().reset_index()
```

# III. Comparing 2016 vs. 2017

```python
#Your Code Here
```

```python
# 1.	Calculate the average run value for each event for both 2016 and 2017
```

```python
Event_Value_16 = RE16.groupby(['event'])['Run_Value'].mean().reset_index()
Event_Value_17 = RE17.groupby(['event'])['Run_Value'].mean().reset_index()
Event_Value_16 = Event_Value_16.rename(columns={'Run_Value':'Run_Value_16'})
Event_Value_17 = Event_Value_17.rename(columns={'Run_Value':'Run_Value_17'})
```

```python
# 2.	Merge the 2016 and 2017 run values by event into one dataframe 
# (before merging, it may be helpful to rename the run value variables Run_Value16 and Run_Value17 respectively)
```

```python
EV_1617 = pd.merge(Event_Value_16, Event_Value_17, on=['event'])
EV_1617
```

```python
# 3.	Calculate the aggregate run value (sum) for all players in both 2016 and 2017
```

```python
Player_Value_16 = RE16.groupby(['batterName'])['Run_Value'].sum().reset_index()
Player_Value_17 = RE17.groupby(['batterName'])['Run_Value'].sum().reset_index()
Player_Value_16 = Player_Value_16.rename(columns={'Run_Value':'Run_Value_16'})
Player_Value_17 = Player_Value_17.rename(columns={'Run_Value':'Run_Value_17'})
```

```python
# 4.	Merge the 2016 and 2017 run values by player into one dataframe so that only players appearing in both datasets appear in the merged data 
# (before merging, it may be helpful to rename the run value variables Run_Value16 and Run_Value17 respectively)
```

```python
PV_1617 = pd.merge(Player_Value_16, Player_Value_17, on=['batterName'])
PV_1617
```

```python
# Q1: What was the highest player run value in 2017?
```

```python
RV2017 = PV_1617.sort_values(by=['Run_Value_17'],  ascending=False)
RV2017
```

```python
# Q2: What was the lowest player run value in 2016?
```

```python
RV2016 = PV_1617.sort_values(by=['Run_Value_16'],  ascending=True)
RV2016
```

```python
# Q3: For each event, calculate the difference in run value between 2017 and 2016 (RV 2017-RV 2016). 
#     Which event saw the largest change (in absolute value) from 2016 to 2017?
```

```python
EV_1617['RVDiff'] = abs(EV_1617['Run_Value_17'] - EV_1617['Run_Value_16']) 
EV_1617 = EV_1617.sort_values(by=['RVDiff'], ascending=False)
EV_1617
```

```python
# Q4: Calculate the difference in player run value between 2017 and 2016 for players that accumulated run values in both seasons. 
# According to this calculation, what was the largest improvement in run value from 2016 to 2017?
```

```python
PV_1617['RVDiff'] = PV_1617['Run_Value_17'] - PV_1617['Run_Value_16']
PV_1617 = PV_1617.sort_values(by=['RVDiff'], ascending=False)
PV_1617
```
