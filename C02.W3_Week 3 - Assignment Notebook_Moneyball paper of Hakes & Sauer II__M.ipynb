# I. Data Preparation

```python
# 2.	Read in “Master.csv” data (created from Week 2)
```

```python
# Read in master data
# Note this master dataset was created in the Week 2 assignment
Master = pd.read_csv("../Data/Master.csv")
```

```python
#Your Code Here
```

```python
# we can see here the names of all the variables in our df
print(Master.columns.tolist())
```

```python
# 3.	Create “Experience Squared” variable
```

```python
Master['Exp2'] = Master['Exp']**2
```

```python
# 4.	Calculate variables for batting average, isolated power, and eye as defined above (for “eye” do not include IBB in walks)
```

```python
# •	Batting Average = Hits/At-Bats
Master['AVG'] = Master['H']/Master['AB']
```

```python
# •	Isolated Power = Slugging Percentage – Batting Average
Master['ISO'] = Master['SLG']-Master['AVG']
```

```python
# •	Eye = (Walks + Hit By Pitches)/Plate Appearances
Master['Eye'] = (Master['BB']+Master['HBP'])/Master['PA']
```

```python
# 5.	Subset the data to only include seasons (SalYear) 1995-2015
```

```python
Master = Master[(Master.SalYear >= 1995) & (Master.SalYear <= 2015)]
display(Master)
```

```python
# Q1: What is the highest single season “Eye” measure for a player across all seasons in the data? 
max_Eye = max(Master['Eye'])

display(max_Eye)
```

```python
# Q2: Calculate the average “ISO” by team for all seasons in the data.  
#     What season does the maximum average “ISO” by team value occur in?
Mean_ISO_Team = Master.groupby(['yearID','teamID'])['ISO'].mean().reset_index()
Max_ISO_Team = Mean_ISO_Team.sort_values(by=['ISO'], ascending=False)

display(Max_ISO_Team)    
```

```python
# Q3: Calculate the median batting average for every season in the data. Which season had the highest median?
Median_AVG_Seas = Master.groupby(['yearID'])['AVG'].median().reset_index()
Median_AVG_Seas = Median_AVG_Seas.sort_values(by=['AVG'], ascending=False)

display(Median_AVG_Seas)    
```

# II. Running Regressions for Each Season

```python
#Your Code Here
```

```python
# 1.	Write a function to run the following regression for every season in the data: 
#     lnSal on batting average, isolated power, eye, plate appearances, experience, experience-squared, and positional fixed effects for free agents only
```

```python
# Write a function to run the Moneyball regression annually for free agents only
def MBExpandFA(Season):
    MB_Seas = Master[(Master.SalYear == Season) & (Master.Free == 1)]
    global lm
    lm = smf.ols(formula = 'lnSal ~ AVG + ISO + Eye + PA + Exp + Exp2 + C(POS)', data=MB_Seas).fit()
    return;
```

```python
# 2.	Create a list to store regression results (note: index will have to be slightly adjusted to account for shorter timeframe)
```

```python
index = 0
lm_Results = [0]
for index in range(1,20):
    lm_Results.append(index)
    index = index + 1
display(lm_Results)
```

```python
# 3.	Run regression for each season in the data and store results
```

```python
Season = 1995
i = 0
while Season < 2015:
    MBExpandFA(Season)
    lm_Results[i] = lm
    i = i + 1
    Season = Season + 1
```

```python
# 4.	Give each regression result a name which is the season it corresponds to
```

```python
Season = 1995
lm_Season = ["1995"]
for Season in range(1996, 2015):
    lm_Season.append(str(Season))
    Season = Season + 1
```

```python
# 5.	Divide results into following eras:
# •	Pre-MB: 1995-2000
# •	Moneyball Era: 2001-2008
# •	Post-MB: 2009-2014
```

```python
Pre_MB = lm_Season[:6]
Pre_MB
```

```python
MB_Period = lm_Season[6:14]
MB_Period
```

```python
Post_MB = lm_Season[14:]
Post_MB
```

```python
# 6.	Generate result tables for each era (should contain regression results for each individual season within each era). 
# Make sure to include the R-squared and number of observations in the regression output
```

```python
# Regression results from 1995-2000
from statsmodels.iolib.summary2 import summary_col
info_dict={'R-squared' : lambda x: f"{x.rsquared:.2f}",
           'No. observations' : lambda x: f"{int(x.nobs):d}"}
PreMB_Out = summary_col([lm_Results[0],lm_Results[1],lm_Results[2],lm_Results[3],lm_Results[4],
                        lm_Results[5]], model_names=Pre_MB, 
                        regressor_order=['AVG','Eye','ISO'],stars=True, info_dict = info_dict)
PreMB_Out
```

```python
# Regression results from 2001-2008
MB_Out = summary_col([lm_Results[6],lm_Results[7],lm_Results[8],lm_Results[9],lm_Results[10],lm_Results[11],
                        lm_Results[12],lm_Results[13]], model_names=MB_Period, 
                        regressor_order=['AVG','Eye','ISO'],stars=True,info_dict = info_dict)
MB_Out
```

```python
# Regression results from 2009-2014
PostMB_Out = summary_col([lm_Results[14],lm_Results[15],lm_Results[16],lm_Results[17],lm_Results[18],
                        lm_Results[19]], model_names=Post_MB, 
                        regressor_order=['AVG','Eye','ISO'],stars=True,info_dict = info_dict)
PostMB_Out
```

```python
# 1.	Subset data to only include free agent players and add a “PreMB” dummy variable to your dataframe for seasons prior to 2004
```

```python
# Create a df made up only of free agents and create a dummy variable 'PreMB'
Master_Free = Master[Master.Free == 1].copy()
Master_Free['PreMB'] = np.where(Master_Free['SalYear']<2004,1,0)
display(Master_Free)
```

```python
# 2.	Run a pooled regression for free agent players using the format same model structure developed in your function in 6). 
# Your pooled regression should have the form: lnSal ~ (model from 6)) + Pre-MB*(model from 6))
```

```python
# Pooled Regression Results
Pooled_lm = smf.ols(formula = 'lnSal ~ AVG + ISO + Eye + PA + Exp + Exp2 + C(POS) + PreMB*(AVG + ISO + Eye + PA + Exp + Exp2 + C(POS))',\
                    data=Master_Free).fit()
Pooled_lm.summary()
```
